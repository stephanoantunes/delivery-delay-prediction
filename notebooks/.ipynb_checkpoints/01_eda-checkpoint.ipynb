# Análise Exploratória de Dados (EDA)

## Projeto: Predição de Atrasos em Entregas no E-commerce

Neste notebook realizamos a análise exploratória dos dados do dataset Olist,
com o objetivo de entender padrões de entrega e identificar fatores associados
a atrasos.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
orders = pd.read_csv('../data/raw/olist_orders_dataset.csv')
customers = pd.read_csv('../data/raw/olist_customers_dataset.csv')
reviews = pd.read_csv('../data/raw/olist_order_reviews_dataset.csv')
```

## Visão geral dos datasets

- **orders**: informações sobre pedidos, datas de compra, envio e entrega
- **customers**: localização do cliente
- **reviews**: avaliações dos clientes após a entrega

```python
#converterndo o campo data para tipo datetime
date_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col])
```

```python
#valida se entrega foi atrasada ou não
orders['entrega_atrasada'] = (
    orders['order_delivered_customer_date'] >
    orders['order_estimated_delivery_date']
)
```

```python
#calcular porcentagem de pedidos atrasado ou nao.
orders['entrega_atrasada'].value_counts(normalize=True) * 100
```

```python
#mostra grafico de pedidos
sns.countplot(x='entrega_atrasada', data=orders)
plt.title('Distribuição de Entregas Atrasadas')
plt.show()
```

```python
#deixando somente pedidos realmente entregues removendo linhas nulas.
orders_entregues = orders.dropna(
    subset=['order_delivered_customer_date']
).copy()
```

```python
#calculo de percentagem de pedidos atrasados que realmente foram entregues
orders_entregues['entrega_atrasada'].value_counts(normalize=True) * 100
```

```python
#juntar as dataframes orders_entregues com a customers
orders_clientes = orders_entregues.merge(
    customers,
    on='customer_id',
    how='left'
)
```

```python
#calculo medio percentual de atraso por estado.
atraso_por_estado = (
    orders_clientes
    .groupby('customer_state')['entrega_atrasada']
    .mean()
    .sort_values(ascending=False)
)

atraso_por_estado.head(10)
```

```python
# grafico percentual de atraso por estado.
plt.figure(figsize=(10,5))
atraso_por_estado.plot(kind='bar')
plt.title('Proporção de Entregas Atrasadas por Estado')
plt.ylabel('Proporção de Atraso')
plt.show()
```

```python
#junta dataframe de pedidos com avalisações dos pedidos.
orders_reviews = orders_clientes.merge(
    reviews[['order_id', 'review_score']],
    on='order_id',
    how='left'
)
```

```python
#calculo de media de avaliações
orders_reviews.groupby('entrega_atrasada')['review_score'].mean()
```

## Principais Insights da Análise Exploratória

- A taxa de entregas atrasadas varia significativamente entre os estados brasileiros.
- Entregas realizadas com atraso apresentam, em média, avaliações mais baixas.
- Isso indica uma relação direta entre desempenho logístico e satisfação do cliente.
